```python


#libreria
import pandas as pd

# Se lee el archivo titanic.csv
df = pd.read_csv('sample_data/titanic.csv')

# Se obtiene estadísticas de los datos
print("--- Estadísticas Generales ---")
print(df.describe())
print("-" * 40)

# Se calcula el porcentaje promedio de la columna 'Survived'
porcentaje_supervivencia = df['Survived'].mean() * 100
print(f"Porcentaje de supervivencia: {porcentaje_supervivencia:.2f}%")

# Se calcula el porcentaje de pasajeros segun cada clase
porcentaje_clase = df['Pclass'].value_counts(normalize=True) * 100
print("\n--- Porcentaje de pasajeros según clase ---")
print(porcentaje_clase)

# Determina cuántos hombres y mujeres viajaron en el Titanic
conteo_genero = df['Sex'].value_counts()
print("\n--- Conteo de pasajeros por género ---")
print(conteo_genero)

print("-" * 40)
print("Proceso Completo")

# A traves de este ejercicio pude poner a prueba la teoria
# presentada en toda la guia de aprendizaje,
# al igual entender de mejor manera lo aprendido en clase, de que los datos
# adquieren una utilidad y se convierten en informacion de valor cuando son
# procesados y darles un contexto, se permite tener una vision clara de las
# estadisticas generales.
# El conocer el entorno de Google Colab ha sido igual algo nuevo y muy
# interesante ya que permite visualizar resultados en el documento dinamico
# sin necesidades de instalaciones locales. Tambien en la guia se pide
# calcular el porcentaje de pasajeros con (Parch) pero lo reemplace
# con (Pclass) ya que al analizar el documento esa es la columna que
# realmente contiene la informacion de las clases del barco, mientras que
# (Parch) representa la cantidad de padres o hijos, lo cual permitio
# que los datos se reflejaran de mejor manera.
```
